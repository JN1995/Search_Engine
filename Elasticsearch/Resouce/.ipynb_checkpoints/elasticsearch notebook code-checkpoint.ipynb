## Create Index
<hr/>

[Create Index](https://www.elastic.co/guide/en/elasticsearch/reference/5.5/indices-create-index.html)
<br/>

1. Index Settings

    Each index created can have specific settings associated with it.

        curl -X PUT "localhost:9200/twitter" -H 'Content-Type: application/json' -d'
        {
            "settings" : {
                "index" : {
                    "number_of_shards" : 5, 
                    "number_of_replicas" : 1 
                }
            }
        }
        '
    ==> Have some other settings, such as: Wait For Active Shards. Its possible values can be found [here](https://www.elastic.co/guide/en/elasticsearch/reference/5.5/docs-index_.html#index-wait-for-active-shards).
    
2. Mappings

    The create index API allows to provide a set of one or more mappings:
    
        curl -X PUT "localhost:9200/test" -H 'Content-Type: application/json' -d'
        {
            "settings" : {
                "number_of_shards" : 5
            },
            "mappings" : {
                "type1" : {
                    "properties" : {
                        "field1" : { "type" : "text" }
                    }
                }
            }
        }
        '
        
3. Aliases

    The create index API allows also to provide a set of [aliases](https://www.elastic.co/guide/en/elasticsearch/reference/5.5/indices-aliases.html):
    
        curl -X PUT "localhost:9200/test" -H 'Content-Type: application/json' -d'
        {
            "aliases" : {
                "alias_1" : {},
                "alias_2" : {
                    "filter" : {
                        "term" : {"user" : "kimchy" }
                    },
                    "routing" : "kimchy"
                }
            }
        }
        '

==> The index creation response will indicate what happened: 
    {
        "acknowledged": true,
        "shards_acknowledged": true
    }

```python
# using with terminal command
# 
!curl -X PUT "localhost:9200/test" -H 'Content-Type: application/json' -d'{"settings" : {"number_of_shards" : 5},"mappings" : {"type1" : {"properties" : {"field1" : { "type" : "text" } } } } }'
```

### With pyelasticsearch module
<hr/>

1. Single Operation

    [create_index method](https://pyelasticsearch.readthedocs.io/en/latest/_modules/pyelasticsearch/client/#ElasticSearch.create_index)
    
    Require: 1 arguments at least (index) which we need to create. <br/>
    This method used to **Create an index with optional setting**.
    Responed: indicate what happened

    [index method](https://pyelasticsearch.readthedocs.io/en/latest/_modules/pyelasticsearch/client/#ElasticSearch.index)
    
    Require: 3 arguments at least -> create index (if index no exists) and document for this index. <br/>
    This method used to **put a typed JSON document into a specific index to make it reseachable**.
    Responed: document is created/updated.

```python
# in pyelasticseach,
# with single operation,
# create variable hold value for settng
_settings = {
                "settings" : {
                    "index" : {
                        "number_of_shards" : 5, 
                        "number_of_replicas" : 1 
                    }
                }
            }
```

```python
print(_settings)
```

```python
# update dict
# _settings.update([("settings", {"index":{"number_of_shards":5}}), ("mappings",1)])
```

```python
# we can use: create_index instance method of ElasticSearch class
pyels.create_index(index='test', settings=_settings)
```

```python
# we can use: index instance method of ElasticSearch class
pyels.index(index='contacts', doc_type='person', overwrite_existing=True, doc='{"name":"David", "age":12}', id=1)
```

## Delete Index
<hr/>

[Delete Index](https://www.elastic.co/guide/en/elasticsearch/reference/5.5/indices-delete-index.html)
<br/>

The delete index API allows to delete an existing index.

    curl -X DELETE "localhost:9200/twitter?pretty=true"
    
The delete index API can also be applied to more than one index, by either using a comma separated list, or on all indices (be careful!) by using _all or * as index.

    curl -X DELETE "localhost:9200/_all"

```python
# using with terminal command
!curl -X DELETE "localhost:9200/twitter?pretty=true"
```

```python
# in pyelasticsearch,
# to delete one/more index -> we use delete_index method.
# we need to pass one/more index to delete
#
pyels.delete_index(index=['contact'])
#
# to delete all index -> we use delete_all_indexes method.
#
# pyels.delete_all_indexes() # -> best
```

## Get Index
<hr/>

[Get Index](https://www.elastic.co/guide/en/elasticsearch/reference/5.5/indices-get-index.html)
<br/>

To retrieve information about one or more indexes.

    curl -X GET "localhost:9200/contacts?pretty=true"

Specifying an index, alias or wildcard expression is required. <br/>
Can also be applied to more than one index, or on all indices by using _all or * as index.

<br/>
Can be filtered to include only specific features by specifying a comma delimited list of features in the URL:
    
    curl -X GET "localhost:9200/contacts/_settings,_mappings?pretty=true"


```python
# using with terminal command
!curl -X GET "localhost:9200/contacts?pretty=true"
```

```python
# variant
!curl -X GET "localhost:9200/contacts/_settings,_mappings?pretty=true"
```

```python
# in pyelasticsearch, we haven't found this method
# we can replace with send_request method
# => responed: result is dictionary type -> 
temp = pyels.send_request(method='GET', path_components=['contacts', '_settings'], query_params={'pretty':'true'})
```

```python
# with dictionary type, we can use json.dumps to format view
# let's discovery this function later.
#
import json

print(json.dumps(temp, indent = 4))
```
